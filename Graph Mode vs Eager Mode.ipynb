```python
import tensorflow as tf
import numpy as np
import time
```

```python
# Eager mode
def penjumlahan_eager_mode(x):
  total = tf.constant(0.0)
  for i in tf.range(100000):
    total += tf.cast(i, dtype=tf.float32) * x
  return total
```

```python
start_time = time.time()
hasil = penjumlahan_eager_mode(tf.constant(2.0))
end_time = time.time()
```

```python
print(f"Hasil Eager Mode: {hasil.numpy()}")
print(f"Waktu Eksekusi Eager Mode: {end_time - start_time:.4f} detik")
```

Menggunakan @tf.function

```python
# Graph mode

@tf.function
def penjumlahan_graph_mode(x):
  total = tf.constant(0.0)
  for i in tf.range(100000):
    total += tf.cast(i, dtype=tf.float32) * x
  return total
```

```python
start_time = time.time()
hasil = penjumlahan_graph_mode(tf.constant(2.0))
end_time = time.time()
```

```python
print(f"Hasil Graph Mode: {hasil.numpy()}")
print(f"Waktu Eksekusi Graph Mode: {end_time - start_time:.4f} detik")
```

```python
print(tf.autograph.to_code(penjumlahan_graph_mode.python_function))
```

Atau menggunakan tf.function

```python
# Graph mode

def penjumlahan_graph_mode_v2(x):
  total = tf.constant(0.0)
  for i in tf.range(100000):
    total += tf.cast(i, dtype=tf.float32) * x
  return total
```

```python
start_time = time.time()
graph_version_function = tf.function(penjumlahan_graph_mode_v2)
hasil = graph_version_function(tf.constant(2.0))
end_time = time.time()
```

```python
print(f"Hasil Graph Mode: {hasil.numpy()}")
print(f"Waktu Eksekusi Graph Mode: {end_time - start_time:.4f} detik")
```

```python
print(tf.autograph.to_code(graph_version_function.python_function))
```
